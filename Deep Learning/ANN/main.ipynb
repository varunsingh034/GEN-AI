```python
import pandas as pd
```

```python
df = pd.read_csv('cust_satisfaction.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df['Gender'] = df['Gender'].map({'Male': 1,'Female' : 0})
df['Customer Type'] = df['Customer Type'].map({'Loyal Customer': 1, 'disloyal Customer': 0})
df['Type of Travel'] = df['Type of Travel'].map({'Business travel': 1, 'Personal Travel': 0})
df['Class'] = df['Class'].map({'Eco': 0, 'Eco Plus': 1, 'Business': 2})
df['satisfaction'] = df['satisfaction'].map({'satisfied': 1, 'neutral or dissatisfied': 0})
df
```

```python
loyal_customer = df[df['Customer Type'] == 1].loc[0:20000]
disloyal_customer = df[df['Customer Type'] == 0]
```

```python
loyal_customer
```

```python
new_data = pd.concat([loyal_customer,disloyal_customer],axis = 0)
new_data
```

```python
x = new_data.drop('Customer Type',axis=1)
y = new_data[['Customer Type']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() # mean = 0 and sd = 1
x_train_scaled = sc.fit_transform(x_train)
# fit == learns the parameter
# transform -- convert the data
x_test_scaled = sc.transform(x_test)
```
